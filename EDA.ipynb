```
# read in data
data = pd.read_csv('creditcard.csv')
```

```
# checking data types
data.info()
```

```
data.head()
```

```
# checking for null data
data.isnull().sum().values
```

```
# checking for duplicated data
data.duplicated().sum()
```

```
# checking for skewness
skew = data.skew().sort_values(ascending = False)
skew_df = pd.DataFrame({'skew': skew})
skew_df
```

```
# checking for kurtosis
kurtosis= pd.DataFrame(data.kurtosis(),columns=['Kurtosis'])
kurtosis
```

```
# checking for variance, eliminate variables with low variance
var = data.var().sort_values(ascending=True )
var_df = pd.DataFrame({'var':var})
var_df
```

```
# checking quantile data
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```
